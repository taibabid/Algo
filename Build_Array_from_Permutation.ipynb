```python
def buildArray(nums) :
        n = len(nums)
        for i in range(0, len(nums)):
            nums[i]=nums[i]+(n*(nums[nums[i]]%n))

        for i in range(0, len(nums)):
            nums[i] = int(nums[i]/n)

        return nums
```

Example 1:

```python
nums = [0,2,1,5,3,4]
#expected result: [0,1,2,4,5,3]
buildArray(nums)
```

Example 2:

```python
nums = [5,0,1,2,3,4]
#expected result: [4,5,0,1,2,3]
buildArray(nums)
```
