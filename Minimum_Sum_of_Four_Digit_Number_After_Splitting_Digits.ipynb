```python
def minimumSum(num):
        nums = sorted(map(int, list(str(num))))
        a = b = 0
        for x in nums:
            a = a*10+x
            a, b = b, a
        return a+b
```

Example 1:

```python
num = 2932
#expected output: 52
minimumSum(num)
```

Example 2:

```python
num = 4009
#expected output: 13
minimumSum(num)
```
